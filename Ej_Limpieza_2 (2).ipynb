```python
import pandas as pd
d = pd.read_csv('Pokemon.csv')
d.head()
```

```python
d.tail()
```

b) Eliminar columnas no deseadas

```python
d.drop(columns='Legendary',inplace=True)
d.drop(columns='Generation',inplace=True)
d.head()
# Eliminamos la columna Legendary y Generation porque no son útiles para lo que ocupamos de la base de datos.
```


e) Crear un nuevo archivo con la base de datos editada

```python
import pandas as pd

d = pd.read_csv('Pokemon.csv')
d.head()

d.drop(columns='Legendary',inplace=True)
d.drop(columns='Generation',inplace=True)
d.head()

d = d.fillna('No hay')
d.head()

d.to_csv('Pokemon_limpio.csv', index=False)
```
